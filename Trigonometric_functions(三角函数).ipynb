## 1. 引入模块

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## 2. 准备数据

```python
x_sin = np.arange(0,4.02*np.pi,0.01*np.pi)   # start,stop,step

x_cos = np.arange(0,4.02*np.pi,0.01*np.pi)   # start,stop,step

sin_fs=(
    lambda x: x,
    lambda x: 2*x,
    lambda x: 0.5*x
)
cos_fs=(
    lambda x: x,
    lambda x: x+0.5*np.pi,
    lambda x: x-0.5*np.pi
)

sin_labels=(
    "$y=\sin(x)$",
    "$y=\sin(2x)$",
    "$y=\sin(\\frac{x}{2})$"
)

cos_labels=(
    "$y=\cos(x)$",
    "$y=\cos(x+\\frac{\\pi}{2})$",
    "$y=\cos(x-\\frac{\\pi}{2})$"
)

c="r"
print(cos_labels[2])
# y = np.sin(x)
# y1 = np.sin(2*x)
# y2 = np.sin(0.5*x)
```

## 3. 绘图    
#### $y=\sin(x)$ : 正常的正弦曲线，周期为 $2\pi$
#### $y=\sin(2x)$ : 沿`x` 轴横坐标缩短 `2` 倍，周期为 $\pi$
#### $y=\sin(\frac{x}{2})$ : 沿 `x`  轴横坐标伸长 `2` 倍，周期为 $4\pi$

#### $y=\cos(x)$  : 正常的余弦曲线，周期为 $2\pi$
#### $y=\cos(x+\frac{\pi}{2})$  : 沿`x` 轴横坐标向左平移$\frac{\pi}{2}$，周期为 $2\pi$
#### $y=\cos(x-\pi)$ : 沿`x` 轴横坐标向左右平移$\frac{\pi}{2}$，周期为 $2\pi$

```python
plt.figure(figsize=(16, 8))
plt.subplot(2, 1, 1)
for sin_f, label in zip(sin_fs, sin_labels):
    y=np.sin(sin_f(x_sin))
    plt.plot(x_sin, y, label=label)
    plt.legend()  

plt.axvline(x=np.pi, c=c, ls="--", lw=1)
plt.axvline(x=2*np.pi, c=c, ls="--", lw=1)

plt.grid()

plt.subplot(2, 1, 2)
for cos_f, label in zip(cos_fs, cos_labels):
    y=np.cos(cos_f(x_cos))
    plt.plot(x_cos, y, label=label)
    plt.legend() 
  
plt.grid()

plt.axvline(x=0.5*np.pi, c=c, ls="--", lw=1)
plt.axvline(x=np.pi, c=c, ls="--", lw=1)
plt.axvline(x=1.5*np.pi, c=c, ls="--", lw=1)
plt.axvline(x=2*np.pi, c=c, ls="--", lw=1)

plt.show()
```

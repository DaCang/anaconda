## 1. 引入模块

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## 2. 准备数据

```python
x = np.arange(0,4.02*np.pi,0.1)   # start,stop,step
sin_fs=(
    lambda x: x,
    lambda x: 2*x,
    lambda x: 0.5*x
)

labels=(
    "$y=\sin(x)$",
    "$y=\sin(2x)$",
    "$y=\sin(\\frac{x}{2})$"
)
# print(labels[2])
# y = np.sin(x)
# y1 = np.sin(2*x)
# y2 = np.sin(0.5*x)
```

## 3. 绘图    
### $y=\sin(x)$    
### $y=\sin(2x)$   
### $y=\sin(\frac{x}{2})$

```python
plt.figure(figsize=(16, 8))

for sin_f, label in zip(sin_fs, labels):
    y=np.sin(sin_f(x))
    plt.plot(x, y, label=label)
    plt.legend()  
plt.axvline(x=4*np.pi, c="#888", ls="--", lw=2)
plt.grid()

plt.show()
```

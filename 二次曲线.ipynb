## 二次曲线画法

### 1. 引入模块

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2. 准备数据 $ x \in R$ 以及 $ y=x^2+2x-1 $

```python
x = np.linspace(-6, 4, 101)
y = x*x+2*x-1

# print(x)
titles = (
    "$y=x^2+2x-1$",
    "$y=x^2+2x-1(x>=-2)$",
    "$y=x^2+2x-1(x=-1)$",
    "$y=x^2+2x-1(x>=2)$",
)
colors=(
    "r","b","y","m"
)

# 右边数据
xflags = (
    lambda arg1: 0<1,
    lambda arg1: arg1>=-2,
    lambda arg2: arg2>=-1,
    lambda arg3: arg3>=2
)

ax_left=[]
ax_right=[]
for xflag in xflags:
    x_left=[]
    x_right=[]
    for xx,yy in zip(x,y):
        if xflag(xx):
            x_right.append(round(xx,2))
        else:
            x_left.append(round(xx,2))
    ax_left.append(x_left)
    ax_right.append(x_right)

Y=np.array(y)

# print(ax_left[1])
# print(ax_right[1])
# print(Y)
# print(len(ax_right[0]))
# print(Y[len(y)-len(ax_right[1]):][:1])
```

### 3. 绘制曲线 

#### 1. 使用 `plt.subplots` 生成 $ 2*2 $ 的图表；

#### 2. **`ax.plot`绘制曲线，`ax.axvline` 绘制辅助线;

#### 3. 先绘制左边部分(`ax_left`)再绘制右边部分(`ax_right`);

```python
# Set-up 2x2 grid for plotting.
fig, sub = plt.subplots(2, 2,figsize=(12, 10))
plt.subplots_adjust(wspace=0.4, hspace=0.4)


for title,color,axx_left,axx_right, ax in zip( titles, colors, ax_left,ax_right, sub.flatten()):
    ax.plot(axx_left,Y[:len(axx_left)],linestyle = 'dotted' ,color ='#888' )
    ax.plot(axx_right,Y[len(y)-len(axx_right):],color =color)
    if axx_right[:1][0]<=-1:
        ax.scatter(-1,-2)
        ax.text(-1, -2, (-1,-2),ha='right', va='top', fontsize=12) 
        ax.axvline(x=-1, c="#888", ls="--", lw=1)
    else:
        ax.scatter(axx_right[:1], Y[len(y)-len(axx_right):][:1])
        ax.text(axx_right[:1][0], Y[len(y)-len(axx_right):][:1][0], (axx_right[:1][0], Y[len(y)-len(axx_right):][:1][0]),
                ha='right', va='top', fontsize=12) 
        ax.axvline(x=axx_right[:1][0], c="#888", ls="--", lw=1)
    ax.set(xlabel="$x$",ylabel="$y$",title= title)
    
```
